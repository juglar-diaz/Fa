```
import os
#import dircache
import codecs
import math
sep = os.sep
```

```
languages = ['spanish','english','portuguese','italian','french','german','dutch','dutch','greek','swedish','danish','finnish']

colecc    = {}
trigramas = {}
trigramas_out = {}

T = 0
TC = 0
TD = 0
TCD = 0
CANT_IDIOMAS_A_REC   = len(languages)
```

```
"""
CARGA EN EL MAPA trigramas EL CONTENIDO DEL FICHERO DADO
""" 
#carga en un map el fichero q contiene el peso de todos los trigramas por idioma
#trabaja con el fichero generado por dar trigramas       
def chargeFileInMap(dirFile=workspace+sep+'ficheros_con_trigramas_pesados_todo'+sep+'frecDocumentos.txt'):
    trigramas.clear()
    f= codecs.open(dirFile,'r','utf-8')
    cont= 0
    for itf in f:
        cont= cont + 1
        
        lf= itf.split()
        aux= len(lf)-1
        trigramas[lf[0]]= []
        
        for idioma in range(aux):
            trigramas[lf[0]].append(float(lf[idioma+1]))
    #print cont                
    global CANT_IDIOMAS_A_REC
    CANT_IDIOMAS_A_REC= aux     
    f.close()
    
    #Verificacion del metodo anterior
    #printMapInFile(workspace+sep+"trigramasfrecDocumentos_test.txt")
```

```
"""
IMPRIME EN EL FICHERO DADO EL CONTENIDO DEL MAPA trigramas
"""
def printMapInFile(dirFile):
    f = codecs.open(dirFile,'w','utf-8')
    for i in trigramas.keys():
        f.write (i)
        for idioma in range(CANT_IDIOMAS_A_REC):
            f.write(' ')
            f.write(str(trigramas[i][idioma]))
        f.write('\n')
    f.close()
```

```
def extracTrig (path):
    #print path
    if os.path.exists(path):
        
        file_out = codecs.open (path+'.trig','w','utf-8')
        file_in = codecs.open (path,'r','utf-8')
        map={}
        
        
        a=''
       
        for it in file_in:
            l = it.split()            
            for itl in l:
            
                if(itl.isalpha()):
                    s='_'+itl+'_'
                    
                    for i in range(0,len(s)-2):
                        a=''
                        for ir in range(0,3):
                            a=a+s[i+ir]
                            a=a.lower()
                        if  not (a in map):
                            map [a]=1
                            
                        else:
                            map[a]=map[a] +1   
                            
        for itout in map.keys():
            file_out.write(itout+' '+str(map[itout])+'\n')
        file_out.close()
        file_in.close()  
```

```
def convertToTrig(dir = workspace+sep+'colecciones'+sep+'europarl_raw_copy_2'):
    l= dircache.listdir(dir)
    for idi in l:
        l1= dircache.listdir(dir+sep+idi)
        for fich_name in l1:
            if (fich_name.count('trig') == 0):
                if(not (fich_name+".trig") in l1):
                    extracTrig( dir + sep +idi+sep+fich_name)
```

```


"""
data ES UNA LISTA Q CONTIENE MAPAS CUYAS LLAVES SON TRIGRAMAS Y EL
CONTENIDO ES LA CANTIDAD DE OCURRENCIAS DE ESE TRIGRAMA PARA EL IDIOMA 
Y LA PARTE ESPECIFICA
"""

def charge(dir = workspace+sep+'colecciones'+sep+'europarl_raw_copy_2'):
    convertToTrig(dir)
    for idioma in range(CANT_IDIOMAS_A_REC):
        idi = languages[idioma]
        colecc[idi] = {}
        l1= dircache.listdir(dir+sep+idi)
        for fich_name in l1:
            if (fich_name.count('trig') == 1):
                fich = codecs.open(dir+sep+idi+sep+fich_name,'r','utf-8')
                
                for line in fich:
                    trig = line.split()[0]
                    if not (trig in colecc[idi]):
                        colecc[idi][trig]=1
                    else: 
                        colecc[idi][trig]=colecc[idi][trig]+int(line.split()[1])
```

```
            
"""
CALCULA LA CANT DE TRIGRAMAS EN LA COLECCION
""" 
def t():
    global T
    cont = 0
    for i in trigramas:
        for j in range(len(trigramas[i])):
            cont += trigramas[i][j]
    T = cont 
    return cont

"""
CALCULA LA CANT DE TRIGRAMAS EN UNA CLASE (idioma)
"""
def tc(idioma):
    global TC
    cont = 0
    for i in trigramas:
        cont += trigramas[i][idioma]
    TC = cont
    return cont

"""
CALCULA LA CANT DE TRIGRAMAS DISTINTOS EN LA COLECCION
"""
def td():
    global TD
    TD = len(trigramas)
    return TD

"""
CALCULA LA CANT DE TRIGRAMAS DISTINTOS EN UNA CLASE (idioma)
"""
def tcd(idioma):
    global TCD
    cont = 0
    for i in trigramas:
        if(trigramas[i][idioma] > 0):
            cont += 1
    TCD = cont
    return cont

"""
CALCULA LA CANT DE OCURRENCIAS DE UN TRIGRAMA EN LA CLASE (IDIOMA)
""" 
def tct(trig,idioma):
    cont = 0
    if(trig in trigramas):
        return trigramas[trig][idioma]
    return cont


def t_c_t(trig,idioma_out):
    cont = 0
    for i in trigramas:
        for j in range(len(trigramas[i])):
            if(trig!=i and j!=idioma_out):
                cont += trigramas[i][j]
    return cont

"""
CALCULA LA CANT DE OCURRENCIAS DE UN TRIGRAMA EN LA COLECCION
""" 
def tt(trig):
    cont = 0
    if(trig in trigramas):
        for j in range(len(trigramas[trig])):
            cont += trigramas[trig][j]
    return cont


"""
CALCULA LA PROBABILIDAD DEL TRIGRAMA EN LA COLECCION
"""
def Pt(trig):
    return float(tt(trig)/float(T)+1.0/T)

def P_t(trig):
    return 1.0-Pt(trig)
    
def Pc(idioma):
    return float(TC)/T
    
def P_c(idioma):
    return 1.0-Pc(idioma)
    
def Ptc(trig,idioma):
    return ((tct(trig,idioma)+1)/float(TC))
    
def P_t_c(trig,idioma):
    return float(t_c_t(trig,idioma))/TC

def P_tc(trig,idioma):
    return (float(tc(idioma))/T)-(Ptc(trig,idioma))

def Pt_c(trig,idioma):
    return (1-float(tc(idioma)))-(P_t_c(trig,idioma))

def frecDocumentos(trig,idioma):
    if (trig in colecc[languages[idioma]]):
        return colecc[languages[idioma]][trig]
    return 0
    #return Ptc(trig,idioma)
    
```

```

def ganInformacion(trig,idioma):
    ptc   = Ptc(trig,idioma)
    pt    = Pt(trig)
    pc    = Pc(idioma)
    
       
    if(pt and pc and ptc):
        return ptc * math.log(float(ptc)/(pt*pc))
    else:
        return 0
```

```

def infMutua(trig,idioma):
    ptc   = Ptc(trig,idioma)
    pt    = Pt(trig)
    pc    = Pc(idioma)
    
    if(pt and pc and ptc):
        return math.log(float(ptc)/(pt*pc))
    else:
        return 0

def chiCuadrado(trig,idioma):
    ptc   = Ptc(trig,idioma)
    p_t_c = P_t_c(trig,idioma)
    p_tc  = P_tc(trig,idioma)
    pt_c  = Pt_c(trig,idioma)
    pt    = Pt(trig)
    pc    = Pc(idioma)
    p_t   = P_t(trig)
    p_c   = P_c(idioma)
    
    num = T * (ptc*p_t_c-pt_c*p_tc)*(ptc*p_t_c-pt_c*p_tc)
    den = pt*pc*p_t*p_c
    if(pt and pc and p_t and p_c ):
        return float(num)/den
    else:
        return 0

```

```
def coRelacion(trig,idioma):
    #print 'cr'
    ptc   = Ptc(trig,idioma)
    p_t_c = P_t_c(trig,idioma)
    p_tc  = P_tc(trig,idioma)
    pt_c  = Pt_c(trig,idioma)
    pt    = Pt(trig)
    pc    = Pc(idioma)
    p_t   = P_t(trig)
    p_c   = P_c(idioma)
    
    num = math.sqrt(T) * (ptc*p_t_c-pt_c*p_tc)
    if(pt and pc and p_t and p_c ):
        den = math.sqrt(pt*pc*p_t*p_c)
        return float(num)/den
    else:
        return 0
```

```
def kullbackLeibler(trig,idioma):
    ptc = Ptc(trig,idioma)
    pt = Pt(trig)
    return ptc * math.log((float(ptc))/(float(pt)))
```
